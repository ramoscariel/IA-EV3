## Carga y exploración inicial del dataset

```python
import pandas as pd

dataset = pd.read_csv('/content/dataset_regresion_multiple_508.csv')
dataset.head()
dataset.describe()
```

## Prepocesamiento

```python
from sklearn.preprocessing import StandardScaler

target_cols = dataset.iloc[:, 1:-2].columns
scaler = StandardScaler()
dataset[target_cols] = scaler.fit_transform(dataset[target_cols])
dataset.head()
```

## División dataset

```python
X = dataset.iloc[:, 1:-2].values
y = dataset.iloc[:, -1].values # Puntaje_Lectura o Puntaje_Matematicas

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Construcción ANN

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(units=64, activation='relu', input_shape=(X_train.shape[1],)),
    keras.layers.Dense(units=64, activation='relu'),
    keras.layers.Dense(units=1)  # Capa de salida con una neurona para regresión lineal
])

# Compila el modelo
model.compile(optimizer='adam', loss='mse')

# Entrena
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

## Predicciones

```python
y_pred = model.predict(X_test)

print(y_pred)
```

## Evaluación

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
```
